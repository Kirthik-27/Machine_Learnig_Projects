```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
```

```python
bcd = sklearn.datasets.load_breast_cancer()
```

```python
print(bcd)
```

```python
df = pd.DataFrame(bcd.data, columns = bcd.feature_names)
```

```python
df['Label'] = bcd.target
```

```python
X = df.drop(columns = 'Label',axis=1)
Y = df['Label']
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2,
                                                    random_state=2)
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
model = keras.Sequential([
                            keras.layers.Flatten(input_shape=(30,)),
                            keras.layers.Dense(30, activation='relu'),#Rectified Linear Unit
                            keras.layers.Dense(30, activation='relu'),
                            keras.layers.Dense(30, activation='relu'),
                            keras.layers.Dense(2, activation='sigmoid')
])
```

```python
model.compile(optimizer = 'adam',
              loss ='sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(X_train, Y_train,validation_split = 0.1,epochs=30)
```

```python
history = model.fit(X_train_std, Y_train,validation_split = 0.1,epochs=50)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['traing data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['traing data', 'validation data'], loc = 'upper right')
```

```python
yloss, accuracy = model.evaluate(X_test_std, Y_test)
```
