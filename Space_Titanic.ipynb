```python
data = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
data['HomePlanet'].fillna(data['HomePlanet'].mode()[0],inplace = True)
```

```python
print(data['HomePlanet'].value_counts())
data.replace({'HomePlanet':{'Earth':1,'Europa':2,'Mars':3},},inplace = True)
data['HomePlanet'].astype('int8')
```

```python

print(data['Destination'].value_counts())
data['Destination'].fillna(data['Destination'].mode()[0],inplace = True)
# print(data['Destination'].value_counts())
data['Destination'] = data['Destination'].map({'TRAPPIST-1e':0, '55 Cancri e':1, 'PSO J318.5-22':2}).astype('int8')
print(data.head())
```

```python
print(data.shape)
data['CryoSleep'].fillna(data['CryoSleep'].mode()[0], inplace = True)
data['CryoSleep'] = data['CryoSleep'].astype('int8')
```

```python
data['VIP'].fillna(data['VIP'].mode()[0], inplace = True)
data['VIP'] = data['VIP'].astype(int)
print(data.head())
data['Cabin'].value_counts()
```

```python
# data[['Cabin_Deck','Cabin_Number','Cabin_Size']] = data['Cabin'].str.split("/",expand = True)
data.drop(columns = ['PassengerId','Name'], axis = 1)
print(data.head())
```

```python
data[['Cabin_Deck','Cabin_Number','Cabin_Side']] = data['Cabin'].str.split('/', expand=True)
```

```python
data.replace({'Cabin_Deck':{'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,'T':8},},inplace = True)
```

```python
data.replace({'Cabin_Side':{'S':1,'P':2},},inplace = True)
print(data.head())
```

```python
data['Cabin_Deck'].fillna(data['Cabin_Deck'].mode()[0], inplace = True)
data['Cabin_Number'].fillna(data['Cabin_Deck'].mode()[0], inplace = True)
data['Cabin_Side'].fillna(data['Cabin_Deck'].mode()[0], inplace = True)
print(data['Cabin_Side'].isna().sum())
print(data['Cabin_Number'].isna().sum())
print(data['Cabin_Deck'].isna().sum())
print(data.head())
```

```python
data['Age'].fillna(data['Age'].mode()[0],inplace=True)
data['FoodCourt'].fillna(data['FoodCourt'].mode()[0],inplace = True)
data['ShoppingMall'].fillna(data['ShoppingMall'].mode()[0],inplace = True)
data['Spa'].fillna(data['Spa'].mode()[0], inplace = True)
data['VRDeck'].fillna(data['VRDeck'].mode()[0], inplace = True)
```

```python
x = data[['HomePlanet','CryoSleep','Destination','Age','VIP','FoodCourt','ShoppingMall','Spa','VRDeck','Cabin_Deck','Cabin_Number','Cabin_Side']]
print(x.isna().sum())
y = data['Transported']
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 2)
print(y.shape)
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
print(y_train.shape)
```

```python
model.fit(x_train, y_train)
```

```python
x_train_predictions = model.predict(x_train)
print(x_train_predictions)
x_test_predictions = model.predict(x_test)
print(x_test_predictions)
```

```python
training_data_accuracy = accuracy_score(y_train,x_train_predictions)
print(training_data_accuracy)
```

```python
test_data_accuracy = accuracy_score(y_test,x_test_predictions)
print(test_data_accuracy)
```
